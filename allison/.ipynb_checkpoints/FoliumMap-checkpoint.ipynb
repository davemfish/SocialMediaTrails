```python
%matplotlib inline
import folium
import geopandas as gpd
import pandas as pd

```

```python
folium.__version__
```

```python
shp = gpd.read_file('../data/pud_results_hex.shp')
```

```python
pud = shp[(shp['PUD_YR_AVG'] > 0)]
```

```python
pud.shape
```

```python
pud.plot()
```

```python
m = folium.Map(location=[47.8, -122.5], zoom_start=7, tiles="cartodbpositron")

m.choropleth(
    geo_data=pud,
    data=pud,
    columns=['FID', 'PUD_YR_AVG'],
    key_on='feature.properties.FID',
    legend_name='PUD_YR_AVG', 
    threshold_scale=[0, 1, 2, 5, 15],
    fill_color='BuPu',
    fill_opacity=0.9,
    highlight=False,
    line_opacity=0)

m.save("photouserdays.html")
```
