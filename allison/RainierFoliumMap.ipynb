```python
%matplotlib inline
import folium
import geopandas as gpd
import pandas as pd
```

```python
folium.__version__
```

```python
shp = gpd.read_file('/Users/peteranderson/Desktop/python/InVEST_OUTPUT/pud_results_18_492_2005_2014.shp')
shp_bdn = gpd.read_file('../data/rainierNP_boundary.shp')
```

```python
pud = shp[(shp['PUD_YR_AVG'] > 0)]
```

```python
pud.plot()
```

```python
m = folium.Map(location=[46.85, -121.76], zoom_start=10, tiles="cartodbpositron")


m.choropleth(
    geo_data=pud,
    data=pud,
    columns=['FID', 'PUD_YR_AVG'],
    key_on='feature.properties.FID',
    legend_name='PUD_YR_AVG', 
    threshold_scale=[0, 1, 2, 5, 15],
    fill_color='BuPu',
    fill_opacity=0.9,
    highlight=False,
    line_opacity=0)

m.save("rainier_photouserdays.html")
```
